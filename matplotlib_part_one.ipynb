```python
# importing the required libraries

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
# creating our datapoints

x = np.arange(0,10)
y = 2*x
```

```python
x
```

```python
y
```

```python
# basic matplotlib plot

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,9)
plt.ylim(0,18)
plt.title('x vs y')
plt.show()

plt.savefig('plot.png',dpi=300)
```
